# ICPC 
###### _Internacional Competitive Programming Contest_

Concurso de universidades de todo el mundo, tiene una duracion por ronda de 5 horas, cada universidad cuenta con 3 a 5 equipos (de principio) de 3 miembros y un solo computador con alrededor de 10 a 15 problemas de todo tipo.

## Etapa Interna.
Se lleva un concurso interno de cada universidad donde saldrán los mejores 3 a 5 equipos de cada universidad.

## Etapa Nacional

Etapa Nacional Bolivia
Es para equipos clasificados en sus universidades
En equipos de 3 personas, 5 horas de concurso y alrededor de 15 problemas.
Clasifican los que mejores ~50 equipos de Bolivia
Etapa Regional, (Bolivia, Chile, Peru, Argentina)
Es para equipos clasificados en sus paises
En equipos de 3 personas, 5 horas de concurso y alrededor de 15 problemas.
Clasifican los mejores 3 de su region.
Etapa Mundial
Es para equipos clasificados en sus regiones
En equipos de 3 personas, 5 horas de concurso y alrededor de 15 problemas.

Punto flotante 
En C++ existen tres tipos de punto flotante que por lo unico que se diferencian es por su capacidad:

float, Este ocupa 4 bytes (32 bits).
double, Este ocupa 8 bytes (64 bits).
long double, Este ocupa 16 bytes (128 bits).
En memoria los puntos flotantes se almacenan en binario pero un poco mas "sofisticados" que los enteros.

El estándar IEEE 754  especifica que un float tiene 1 bit para el signo, 8 bits para el exponente y 24 bits para la fracción, de los cuales solo 23 bits son almacenados en memoria. Un long double tiene 1 bit para el signo, 15 bits para el exponente y 113 bits para la fracción, de los cuales solo 112 bits son almacenados en memoria.
